```python
def HammingDistance(p,q):
    
    l = len(p)
    i = 0
    
    hmd = 0
    
    while i < l :
        if(p[i] != q[i]):
            hmd = hmd + 1
        i = i + 1
        
    return hmd

def minHammingDistance(dna,p):
    pattern = []
    
    i = 0
    while i < len(dna) - len(p) + 1:
        
        tmp = dna[i:i+len(p)]
        #print tmp
        pattern.append(tmp)
        i+=1
    
    #print(*pattern)
    
    i = 0
    minHmd = 99999999999999
    while i < len(pattern):
        hmd = HammingDistance(pattern[i],p)
        if hmd < minHmd:
            minHmd = hmd
            
        i+=1
    
    return minHmd
    
import itertools

DNA = ['AAATTGACGCAT',
'GACGACCACGTT',
'CGTCAGCGCCTG',
'GCTGAGCACCGG',
'AGTACGGGACAG']

K = 3
lst = list(map(''.join, itertools.product('ACGT', repeat=K)))
#print (*lst)

minScore = 99999999999999
minKmer = ''

i = 0
while i < len(lst):
    
    j = 0
    score = 0
    while j < len(DNA):
        score += minHammingDistance(DNA[j],lst[i])
        j+=1
    #print (lst[i] + " ==> " + str(score))
    
    if score < minScore:
        minScore = score
        minKmer = lst[i]
    
    i+=1

print (minKmer)
```
